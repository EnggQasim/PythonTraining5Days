# Apply function on multiple columns and return mutliple columns

```python
def obtained_marks(s1,s2,s3,s4,s5):
    obtained = s1 + s2 + s3 + s4 + s5
    per = (obtained/500)*100
    grade = "Waiting"
    
    if per >= 80:
        grade = "A+"
    elif per >=70:
        grade = "A"
    elif per >= 60:
        grade = "B"
    elif per >= 50:
        grade = "C"
    else:
        grade = "Fail"
    return obtained , per , grade

obtained_marks(1,2,3,4,5)
```

```python
df[['Obtained','Percentage','Grade']] = df[["A","B","C","D","E"]].apply(lambda x:obtained_marks(*x),axis=1, result_type='expand')
df
```
